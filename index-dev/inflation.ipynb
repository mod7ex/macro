```python
from matplotlib import pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
# CPI
CPIAUCSL_df = pd.read_csv('../data/CPIAUCSL.csv')
CPILFESL_df = pd.read_csv('../data/CPILFESL.csv')

# PPI
WPSFD4131_df = pd.read_csv('../data/WPSFD4131.csv')
WPSFD49207_df = pd.read_csv('../data/WPSFD49207.csv')

# PCE
PCEPI_df = pd.read_csv('../data/PCEPI.csv')
PCEPILFE_df = pd.read_csv('../data/PCEPILFE.csv')

inflation_dfs = [CPIAUCSL_df, CPILFESL_df, WPSFD4131_df, WPSFD49207_df, PCEPI_df, PCEPILFE_df]

for df in inflation_dfs:
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.set_index('DATE', inplace=True)
    df.dropna(inplace=True)
```

```python
WINDOW = 60

for df in inflation_dfs:
    key = df.columns[0]
    df[f"rolling_mean({key})"] = df[df.columns[0]].rolling(window=WINDOW).mean()
    df[f"rolling_std({key})"] = df[df.columns[0]].rolling(window=WINDOW).std()
    df[f"rolling_z({key})"] = (df[df.columns[0]] - df[f"rolling_mean({key})"]) / df[f"rolling_std({key})"]
    df.dropna(inplace=True)
    # df = df.index.to_period('M')
```
