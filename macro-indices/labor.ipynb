```python
import sys
import os

sys.path.append(os.path.abspath(".."))
```

```python
from utils import load_report_data, plot_df_chart, describe_data, get_frequency_table
```

```python
DICT = {
    "PAYEMS": {
        "title": "All Employees, Total Nonfarm (NFP)",
        "unit": "Thousands of Persons"
    },
    "UNRATE": {
        "title": "Unemployment Rate",
        "unit": "%"
    },
}
```

```python
INDICATOR = 'UNRATE'

MoM_col = 'MoM ^ %'
YoY_col = 'YoY ^ %'

COL = YoY_col
```

```python
df = load_report_data(INDICATOR)

# df[COL] = df[INDICATOR].pct_change(fill_method=None, periods=12 if COL == YoY_col else 1) * 100

df[COL] = df[INDICATOR].ffill().pct_change(periods=12 if COL == YoY_col else 1) * 100

df.tail()
```

- **Plot of Data**

```python
plot_df_chart(
    df[[INDICATOR]],
    chart_title=DICT[INDICATOR]['title'],
    yaxis_title=DICT[INDICATOR]['unit'],
    use_markers=False
)
```

- **Stats**

```python
df_stats, stats = describe_data(df[COL].dropna())

df_stats
```

```python
freq_table = get_frequency_table(df.dropna()[COL])

freq_table
```

```python
fig = plot_df_chart(
    freq_table[['Probability %']],
    chart_title = f'{DICT[INDICATOR]['title']} (% change frequency)',
    chart_type = "bar",
    yaxis_title = "Probability %",
    width=1000,
    height=700,
    show_rangeslider=False
)
```

```python
import plotly.graph_objects as go

fig = plot_df_chart(
    df[[COL]],
    chart_title = DICT[INDICATOR]["title"] + f' ({COL})',
    chart_type = "bar",
    yaxis_title = "%",
    draw=False,
)

AVG_PERIOD = 12 * 10

fig.add_trace(go.Scatter(
    x=df.index,
    y=df[COL].rolling(window=AVG_PERIOD).mean(),
    name='Average ' + COL + f' ({AVG_PERIOD} months)',
    line=dict(
        width=1,
        color="#F2C14E",
        # dash="dash"
    ),
    # hoverinfo="skip",
    # showlegend=False,
))

fig.show()
```
