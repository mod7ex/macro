```python
DICT = {
    "PERMIT": {
        "title": "New Privately-Owned Housing Units Authorized in Permit-Issuing Places: Total Units",
        "unit": "Thousands of Units"
    }
}
```

```python
INDICATOR = "PERMIT"

MoM_col = 'MoM ^ %'
YoY_col = 'YoY ^ %'

COL = YoY_col
```

- **Load Data**

```python
df = load_report_data(INDICATOR)

# df[COL] = df[INDICATOR].pct_change(fill_method=None, periods=12 if COL == YoY_col else 1) * 100
df[COL] = df[INDICATOR].ffill().pct_change(periods=12 if COL == YoY_col else 1) * 100

df.dropna(inplace=True)

df
```

- **Data plot**

```python
fig = plot_df_chart(
    df[[INDICATOR]],
    chart_title = DICT[INDICATOR]["title"],
    chart_type = "line",
    yaxis_title = "%",
    use_markers=False
)
```

- **% change**

```python
fig = plot_df_chart(
    df[[COL]],
    chart_title = DICT[INDICATOR]["title"] + f' ({COL})',
    chart_type = "bar",
    yaxis_title = "%",
)
```
