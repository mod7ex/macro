```python
import pandas as pd
import plotly.express as px
```

```python
FEDFUNDS_df = pd.read_csv('data/FEDFUNDS.csv')

# FEDFUNDS_df['DATE'] = pd.to_datetime(FEDFUNDS_df['DATE'])

FEDFUNDS_df.set_index('DATE', inplace=True)

FEDFUNDS_df
```

- Plot

```python
fig = px.line(
    FEDFUNDS_df,
    x=FEDFUNDS_df.index,
    y="FEDFUNDS",
    markers=True
)

fig.update_layout(
    title=dict(text='Federal Funds Effective Rate'),
    width=1500,
    height=600,
    yaxis=dict(title="%"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```

- **Fed Funds rate % change**

```python
FEDFUNDS_df['^ %'] = FEDFUNDS_df['FEDFUNDS'].pct_change() * 100
```

```python
fig = px.line(
    FEDFUNDS_df,
    x=FEDFUNDS_df.index,
    y="^ %",
    markers=True
)

fig.update_layout(
    title=dict(text='Fed Funds rate % change'),
    width=1500,
    height=600,
    yaxis=dict(title="%"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```
