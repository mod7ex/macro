```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('data/WGS10YR.csv')

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
```

```python
fig = px.area(
    df,
    x=df.index,
    y="WGS10YR",
)

fig.update_layout(
    title=dict(text='Market Yield on U.S. Treasury Securities at 10-Year Constant Maturity, Quoted on an Investment Basis'),
    width=1500,
    height=600,
    yaxis=dict(title="%"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```

- **T10% % change**

```python
df['^ %'] = 100 * df['WGS10YR'].pct_change()
```

```python
fig = px.line(
    df,
    x=df.index,
    y="^ %",
    markers=True
)

fig.update_layout(
    title=dict(text='T10% % change'),
    width=1500,
    height=600,
    yaxis=dict(title="%"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```
