```python
import pandas as pd
import plotly.graph_objects as go
from utils import DATA_FOLDER_PATH, describe_data, get_frequency_table, plot_df_chart
```

```python
DICT = {
    "CPIAUCSL": {
        "title": "Consumer Price Index for All Urban Consumers: All Items in U.S. City Average",
        "unit": ""
    },
    "CPILFESL": {
        "title": "Consumer Price Index for All Urban Consumers: All Items Less Food and Energy in U.S. City Average",
        "unit": ""
    },
    "WPSFD49207": {
        "title": "Producer Price Index by Commodity: Final Demand: Finished Goods",
        "unit": ""
    },
    "WPSFD4131": {
        "title": "Producer Price Index by Commodity: Final Demand: Finished Goods Less Foods and Energy",
        "unit": ""
    },
    "PCEPILFE": {
        "title": "Personal Consumption Expenditures Excluding Food and Energy (Chain-Type Price Index)",
        "unit": ""
    },
    "PCEPI": {
        "title": "Personal Consumption Expenditures: Chain-type Price Index",
        "unit": ""
    },
}
```

```python
INDICATOR = 'WPSFD49207'

MoM_col = 'MoM ^ %'
YoY_col = 'YoY ^ %'

COL = YoY_col
```

```python
df = pd.read_csv(os.path.join(DATA_FOLDER_PATH, INDICATOR + '.csv'))


df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)

# df[COL] = df[INDICATOR].pct_change(fill_method=None, periods=12 if COL == YoY_col else 1) * 100

df[COL] = df[INDICATOR].ffill().pct_change(periods=12 if COL == YoY_col else 1) * 100

df.tail()
```

- **Stats**

```python
df_stats, stats = describe_data(df[COL].dropna())

df_stats
```

- **Freq**

```python
freq_table = get_frequency_table(df.dropna()[COL])

freq_table
```

```python
fig = plot_df_chart(
    freq_table[['Probability %']],
    chart_title = f'{DICT[INDICATOR]['title']} (% change frequency)',
    chart_type = "bar",
    yaxis_title = "Probability %",
    width=1000,
    height=700,
    show_rangeslider=False
)
```

```python
fig = plot_df_chart(
    df[[COL]],
    chart_title = DICT[INDICATOR]["title"] + f' ({COL})',
    chart_type = "bar",
    yaxis_title = "%",
    draw=False,
)

AVG_PERIOD = 12 * 5

fig.add_trace(go.Scatter(
    x=df.index,
    y=df[COL].rolling(window=AVG_PERIOD).mean(),
    name='Average ' + COL + f' ({AVG_PERIOD} months)',
    line=dict(
        width=1,
        color="#F2C14E",
        # dash="dash"
    ),
    # hoverinfo="skip",
    # showlegend=False,
))

fig.show()
```
