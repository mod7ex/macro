```python
import sys
import os

sys.path.append(os.path.abspath(".."))
```

```python
import pandas as pd
import plotly.graph_objects as go
from utils import DATA_FOLDER_PATH, plot_df_chart
```

```python
DICT = {
    "PMI": {
        "title": "ISM Manufacturing Purchasing Managers Index",
        "unit": "%"
    },
    "NMI": {
        "title": "ISM Non-Manufacturing Purchasing Managers Index",
        "unit": "%"
    }
}
```

```python
INDICATOR = "NMI"
```

```python
df = pd.read_csv(os.path.join(DATA_FOLDER_PATH, 'ISM-' +  INDICATOR + '.csv'))

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)

df
```

- **PMI index chart**

```python
fig = plot_df_chart(
    df,
    chart_title = DICT[INDICATOR]["title"],
    chart_type = "line",
    yaxis_title = DICT[INDICATOR]["unit"],
    draw=False
)

# Reference line
fig.add_trace(go.Scatter(
    x=df.index,
    y=[50]*len(df),
    name='50 Threshold',
    line=dict(
        width=1,
        color="#FFC53D",   # muted gold
    ),
    hoverinfo="skip",
))

fig.show()
```
