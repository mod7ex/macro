```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('data/WALCL.csv')
realGDP_df = pd.read_csv('data/GDPC1.csv')

df['DATE'] = pd.to_datetime(df['DATE'])
realGDP_df['DATE'] = pd.to_datetime(realGDP_df['DATE'])

df.set_index('DATE', inplace=True)
realGDP_df.set_index('DATE', inplace=True)
```

```python
df = df[df.index.month.isin([1, 4, 7, 10])] # keep only quarterly data

df=df.groupby(df.index.to_period("M")).last() # pick last observation of each quarter

df
```

```python
realGDP_df.index = realGDP_df.index.to_period('M')

realGDP_df
```

```python
df = df.join(realGDP_df, how="inner")
```

### Plotting data

```python
df['CBBS % GDP'] = (df['WALCL'] / (df['GDPC1'] * 1000)) * 100

df
```

```python
fig = px.area(
    df,
    x=df.index.to_timestamp(),
    y="CBBS % GDP",
)

fig.update_layout(
    title=dict(text='Central Bank (Fed) Balance Sheet as % of Real GDP', x=0.5),
    width=1000,
    height=600,
    yaxis=dict(title="%"),
    xaxis=dict(title="Date"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```
