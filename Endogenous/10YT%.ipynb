```python
from matplotlib import pyplot as plt
import plotly.graph_objects as go
import pandas as pd

plt.style.use('Solarize_Light2')
```

```python
df = pd.read_csv('data/WGS10YR.csv')

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
```

```python
fig = go.Figure(go.Scatter(x = df.index, y = df['WGS10YR'], fill='tozeroy', name='WGS10YR'))

fig.update_layout(
    title=dict(text='Market Yield on U.S. Treasury Securities at 10-Year Constant Maturity, Quoted on an Investment Basis'),
    # yaxis=dict(title="Percent", exponentformat="none"),
    plot_bgcolor='rgb(230, 230,230)',
    yaxis_title='% Percent',
    width=1400,
    height=500,
    hovermode="x unified",
)

fig.show()
```

- **10Y T% Pct change**

```python
df['^%'] = 100 * df['WGS10YR'].pct_change()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df['^%'],
        mode='lines',
        line=dict(color='#ff3961', width=.8),
        name="10Y T% pct change",
        hovertemplate="<br>Date: %{x|%d-%b-%Y}<br>Value: %{y:.2f}%",
    )
)

fig.update_layout(
    title=dict(text='10Y T% pct change'),
    yaxis_title='Percent',
    width=1400,
    height=700,
    plot_bgcolor="black",
    paper_bgcolor="#212121",
    font=dict(color="white"),
    hovermode="x unified",
    xaxis=dict(
        title="",
        showgrid=True,
        gridcolor="gray",
        tickformat="%Y",
    ),
    yaxis=dict(
        title="% Change",
        showgrid=True,
        gridcolor="gray"
    ),
)

fig.show()
```
