```python
import plotly.graph_objects as go
import pandas as pd
import sys
sys.path.append('../utils')

from helpers import real_gdp_df
```

## Loading Real [GDP](https://fred.stlouisfed.org/series/GDP) data

```python
GDP_df = pd.read_csv('data/GDP.csv')

GDP_df['DATE'] = pd.to_datetime(GDP_df['DATE'])
GDP_df.set_index('DATE', inplace=True)
```

## Data Preview

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=GDP_df.index,
    y=GDP_df['GDP'],
    name='GDP',
    fill='tonexty',
    fillcolor='rgba(0, 0, 255, 0.7)',
    line=dict(color='blue', width=1),
))

fig.update_layout(
    title=dict(text='GDP (Frequency: Quarterly)'),
    plot_bgcolor='rgb(230, 230,230)',
    width=1400,
    height=700,
    hovermode="x unified",
    yaxis=dict(title="Billions of Chained 2017 Dollars", exponentformat="none", showgrid=True),
    barmode="group",
    legend=dict(x=0.5, y=-0.2, orientation="h"),
    xaxis=dict(showgrid=True),
)

fig.show()
```

# Real GDP cpt change

```python
real_GDP_df = real_gdp_df()
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=real_GDP_df.index,
    y=real_GDP_df['GDP'],
    name='Real GDP',
    marker=dict(color='blue'),
))

fig.update_layout(
    title=dict(text='Real GDP (Frequency: Quarterly)'),
    plot_bgcolor='rgb(230, 230,230)',
    width=1400,
    height=700,
    hovermode="x unified",
    yaxis=dict(title="Percent Change from Preceding Period %", exponentformat="none", showgrid=True),
    barmode="group",
    legend=dict(x=0.5, y=-0.2, orientation="h"),
    xaxis=dict(showgrid=True),
)

fig.show()
```
