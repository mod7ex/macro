## Loading Real [GDP](https://fred.stlouisfed.org/series/GDP) data

```python
from matplotlib import pyplot as plt
import pandas as pd
# import sys

# plt.style.use('dark_background')
plt.style.use('Solarize_Light2')

# sys.path.append('../utils')
```

```python
GDP_df = pd.read_csv('data/gdp/GDP.csv')

GDP_df['DATE'] = pd.to_datetime(GDP_df['DATE'])
GDP_df.set_index('DATE', inplace=True)
# GDP_df['DATE'] = pd.to_datetime(GDP_df['DATE']).dt.strftime("%d-%b-%Y")

GDP_df['^ %'] = GDP_df['GDP'].pct_change() * 100
GDP_df
```

## Data Preview

```python
GDP_df['GDP'].plot(
    figsize=(20, 7),
    xlabel='',
    ylabel='Billions of Chained 2017 Dollars',
    title='Real GDP (Frequency: Quarterly)',
);
```

# Real GDP cpt change

```python
real_GDP_df = pd.read_csv('data/A191RL1Q225SBEA.csv')

real_GDP_df['DATE'] = pd.to_datetime(real_GDP_df['DATE'])
real_GDP_df.set_index('DATE', inplace=True)
real_GDP_df.columns = ['Real GDP']

real_GDP_df
```

```python
real_GDP_df['Real GDP'].plot(
    figsize=(17, 5),
    xlabel='',
    ylabel='Percent Change from Preceding Period,',
    title='Real GDP (Frequency: Quarterly)',
);
```

## Real GDP pct change (last 36 months)

```python
real_GDP_df.tail(36)['Real GDP'].plot(
    figsize=(20, 5),
    xlabel='',
    ylabel='Percent Change from Preceding Period,',
    title='Real GDP (Frequency: Quarterly)',
);
```
