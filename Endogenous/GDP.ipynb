## Loading [GDP](https://fred.stlouisfed.org/series/GDP) data

```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import pandas as pd

# plt.style.use('dark_background')
plt.style.use('Solarize_Light2')

# import sys

# sys.path.append('../utils')
```

```python
# plt.style.available
```

```python
GDP_df = pd.read_csv('data/gdp/GDP.csv')

GDP_df['DATE'] = pd.to_datetime(GDP_df['DATE'])
GDP_df.set_index('DATE', inplace=True)
# GDP_df['DATE'] = pd.to_datetime(GDP_df['DATE']).dt.strftime("%d-%b-%Y")

GDP_df['^GDP %'] = GDP_df['GDP'].pct_change() * 100
GDP_df
```

## Data Preview

```python
GDP_df['GDP'].plot(
    figsize=(20, 7),
    xlabel='',
    ylabel='Billions of Dollars',
    title='GDP (Frequency: Quarterly)',
);
```
