```python
import pandas as pd
import plotly.express as px
```

```python
GFDEBTN_df = pd.read_csv('../data/GFDEBTN.csv')
realGDP_df = pd.read_csv('../data/GDPC1.csv')

# GFDEBTN_df['DATE'] = pd.to_datetime(GFDEBTN_df['DATE'])
# realGDP_df['DATE'] = pd.to_datetime(realGDP_df['DATE'])

GFDEBTN_df.set_index('DATE', inplace=True)
realGDP_df.set_index('DATE', inplace=True)
```

```python
df = GFDEBTN_df.join(realGDP_df, how="inner")

df['Debt / GDP Ratio'] = (df['GFDEBTN'] / (df['GDPC1'] * 1000)) * 100

df
```

- **Plot Debt / GDP Ratio**

```python
fig = px.area(
    df,
    x=df.index,
    y="Debt / GDP Ratio",
)

fig.update_layout(
    title=dict(text='Debt / GDP Ratio'),
    width=1500,
    height=600,
    yaxis=dict(title="%"),
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```
