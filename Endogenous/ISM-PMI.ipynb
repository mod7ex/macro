```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df = pd.read_csv('data/ISM-PMI.csv')

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

- **ISM PMI index chart**

```python
mid_chart = go.Scatter(
    x = df.index, 
    y=[50 for i in range(len(df))], 
    mode='lines', 
    name='50 Points',
    line=dict(color='blue', width=1),
)

pmi_chart = go.Scatter(
    x = df.index,
    y = df['Actual'],
    fill='tonexty',
    name='PMI',
    fillcolor='rgba(0, 4, 255, 0.3)',
    line=dict(color='blue', width=1),
)

fig = go.Figure(data=[mid_chart, pmi_chart])

fig.update_layout(
    title=dict(text='ISM PMI index'),
    # yaxis=dict(title="Percent", exponentformat="none"),
    plot_bgcolor='rgb(230, 230,230)',
    yaxis_title='points',
    width=1400,
    height=700,
    hovermode="y unified",
)

fig.show()
```
