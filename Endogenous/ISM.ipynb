```python
import plotly.graph_objects as go
import pandas as pd
```

# **ISM PMI**

```python
pmi_df = pd.read_csv('data/ISM-PMI.csv')

pmi_df['DATE'] = pd.to_datetime(pmi_df['DATE'])
pmi_df.set_index('DATE', inplace=True)

pmi_df
```

- **PMI index chart**

```python
fig = go.Figure()

# Reference line
fig.add_trace(go.Scatter(
    x=pmi_df.index,
    y=[50]*len(pmi_df),
    name='50 Threshold',
    line=dict(
        width=2,
        color="#725615",   # muted gold
        dash="dash"
    ),
    hoverinfo="skip",
))

fig.add_trace(go.Scatter(
    x=pmi_df.index,
    y=pmi_df["ISM(PMI)"],
    name='ISM(PMI)',
    mode='lines+markers',
    line=dict(
        color="#4C78A8",
        width=2
    ),
    marker=dict(
        color="#4C78A8",
        size=6
    )
))


fig.update_layout(
    title=dict(
        text='ISM(PMI)',
        x=0.02,
        font=dict(size=20)
    ),
    plot_bgcolor="#0E1117",
    paper_bgcolor="#0E1117",
    font=dict(color="#E6E6E6", size=14),
    width=1400,
    height=700,
    yaxis=dict(
        title="%",
        exponentformat="none",
        showgrid=False,
        zeroline=False
    ),
    xaxis=dict(
        showgrid=False,
        zeroline=False
    ),
    hovermode="x unified",
)

fig.show()
```

```python
fig = go.Figure()

# Reference line
fig.add_trace(go.Scatter(
    x=pmi_df.index,
    y=[50]*len(pmi_df),
    name='50 Threshold',
    line=dict(
        width=2,
        color="#F2C14E",   # muted gold
        dash="dash"
    ),
    hoverinfo="skip",
))

# Bars
fig.add_trace(go.Bar(
    x=pmi_df.index,
    y=pmi_df["ISM(PMI)"],
    name='ISM(PMI)',
    marker=dict(
        color="#4C78A8",  # professional blue
        line=dict(color="#4C78A8", width=0)
    ),
    width=0.5,
))

fig.update_layout(
    title=dict(
        text='ISM(PMI)',
        x=0.02,
        font=dict(size=20)
    ),
    plot_bgcolor="#0E1117",
    paper_bgcolor="#0E1117",
    font=dict(color="#E6E6E6", size=14),
    width=1400,
    height=700,
    yaxis=dict(
        title="%",
        exponentformat="none",
        showgrid=False,
        zeroline=False
    ),
    xaxis=dict(
        showgrid=False,
        zeroline=False
    ),
    hovermode="x unified",
)

fig.show()
```
